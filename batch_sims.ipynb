```python
import pickle
import os
import glob
```

```python
from sim_runner import baseline_choice, simulate, get_total_conversions, get_regret
```

```python
# Testing the function simulate
num_rounds = 10
choice_function = baseline_choice
total_budget = 170
min_bid, max_bid = .1, 1.

sim_folder = 'sims'
path_list = glob.glob(os.path.join(sim_folder, '*'))

result_folder = 'results_baseline'

for sim_num, path in enumerate(path_list):
    name_hash = os.path.basename(path).replace('.pkl', '')
    done_list = glob.glob(os.path.join(result_folder, '*'))
    if os.path.join(result_folder, name_hash + '.pkl') in done_list:
        continue
    with open(path, 'rb') as file:
        sim = pickle.load(file)
    simulate(sim, choice_function, total_budget, num_rounds, min_bid, max_bid, result_folder, name_hash)
```

```python
# Testing the function get_regret
# Using two different simulations just for the sake of testing 

result_path_list = glob.glob(os.path.join(result_folder, '*'))

with open(result_path_list[0], 'rb') as file:
    sim = pickle.load(file)
    
with open(result_path_list[1], 'rb') as file:
    sim2 = pickle.load(file)
    
get_regret(sim, sim2)
```
